```python
#do MDS using a force field
def getForces(X,V,D,k=1.,gamma=0.):
    nP = len(X)
    F = np.zeros((nP,2))
    W = 0
    
    for i in xrange(nP):
        for j in xrange(i+1,nP):
            difference = X[j] - X[i]            
            distance = np.linalg.norm(difference)
            if distance == 0:
                continue
            direction = difference/distance
            
            f = k* (distance - D[i,j]) * direction
            F[i] = F[i] + f - gamma*V[i]
            F[j] = F[j] - f - gamma*V[j]
            W = W + 0.5*k*(distance - D[i,j])**2.
            
    return F,W
```

```python
def doMinimization(X0,D,k=1.,dt=1e-2,maxsteps=1000,epsilon=1e-3):
    #minimize
    X = X0
    V = []
    for t in xrange(maxsteps):
        f, v = getForces(X,np.zeros(X.shape),D,k=k, gamma=0.)
        X = X + dt*f
        V.append(v)
        
        if t%10:
            print 'max-force: %f, energy: %f' %(np.max(f), v)
        
        if t>2 and V[-2]/v-1<epsilon:
            print 'Terminated after %i steps.' %t
            return X, V
        
    return X, V
```

```python
def doLeapFrog(X0,D,M,k=1,gamma=0.,dt=1e-2,steps=1e3,dtout=0):
    X = np.copy(X0)
    V = np.zeros(X.shape)
    _, w = getForces(X,V,D,k=k,gamma=gamma)
    W = [w]
    X_out = [X0]
    
    for t in xrange(steps):
        F, w = getForces(X,V,D,k=k,gamma=gamma)
        X = X + V * dt
        V = V + (F.T/M).T * dt
        
        if dtout > 0 and not t%dtout:
            X_out.append(np.copy(X))
            W.append(w)
    
    if dtout > 0:       
        return X_out, W
    else:
        return X, w
```
